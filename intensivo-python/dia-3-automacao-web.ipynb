```python
# Bibliotecas
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
# Criação de webdriver
driver = webdriver.Chrome()

# Passo 1: Adquirir as cotações atualizadas
cotacao = []

# Cotação do dólar
driver.get('https://www.google.com')
search_input = driver.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div[2]/div[2]/input')

search_input.send_keys('Cotação do dólar')
search_input.send_keys(Keys.ENTER)

cotacao.append(driver.find_element(By.XPATH,
                                        '//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value'))

# Cotação do euro
driver.get('https://www.google.com')
search_input = driver.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div[2]/div[2]/input')

search_input.send_keys('Cotação do euro')
search_input.send_keys(Keys.ENTER)

cotacao.append(driver.find_element(By.XPATH,
                                        '//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value'))

# Cotação do ouro
driver.get('https://www.melhorcambio.com/ouro-hoje')
cotacao.append(driver.find_element(By.XPATH, '//*[@id="comercial"]').get_attribute('value'))
cotacao[2] = cotacao[2].replace(',','.')

# Finalizar navegador
driver.quit()
```

```python
print(cotacao)
```

```python
# Passo 2: Obter a base de dados
db = pd.read_excel('Produtos.xlsx')
display(db)
```

```python
# Passo 3: Atualizar a base de dados(Cotações e preços)
# Atualização da cotação
db.loc[db['Moeda']=='Dólar','Cotação']=float(cotacao[0])
db.loc[db['Moeda']=='Euro','Cotação']=float(cotacao[1])
db.loc[db['Moeda']=='Ouro','Cotação']=float(cotacao[2])

display(db)
```

```python
# Atualização de preço de compra
db['Preço de Compra'] = db['Preço Original'] * db['Cotação']
display(db)
```

```python
# Atualização de preço de venda
db['Preço de Venda'] = db['Preço de Compra'] * db['Margem']
display(db)
```

```python
# Passo 4: Exportar para uma nova base
db.to_excel('Produtos_atualizado.xlsx', index=False)
```
